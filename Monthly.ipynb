```python
import pandas as pd
import numpy as np

index = pd.read_excel('ENSO.xlsx')
anomaly = pd.read_csv('WWVcleaned.csv')
print(index)
print(anomaly)
```

```python
import pandas as pd
import matplotlib.pyplot as plt



plt.plot(index['Year'], index['DJF'], label = 'DJF ENSO index', color = 'black')
plt.plot(index['Year'], anomaly['NDJ'], label = 'NDJ WWV Anomaly')
plt.plot(index['Year'], anomaly['OND'], label = 'OND WWV Anomaly')
plt.plot(index['Year'], anomaly['SON'], label = 'SON WWV Anomaly')
plt.plot(index['Year'], anomaly['ASO'], label = 'ASO WWV Anomaly')
plt.plot(index['Year'], anomaly['JAS'], label = 'JAS WWV Anomaly')
plt.plot(index['Year'], anomaly['JJA'], label = 'JJA WWV Anomaly')

plt.xlabel('Years')
plt.ylabel('Value')
plt.title('DJF ENSO index correlation')
plt.legend()

plt.show()

cc_NDJ = index['DJF'].corr(anomaly['NDJ'])
cc_OND = index['DJF'].corr(anomaly['OND'])
cc_SON = index['DJF'].corr(anomaly['SON'])
cc_ASO = index['DJF'].corr(anomaly['ASO'])
cc_JAS = index['DJF'].corr(anomaly['JAS'])
cc_JJA = index['DJF'].corr(anomaly['JJA'])

print('Correlation Coefficient of NDJ:', cc_NDJ)
print('Correlation Coefficient of OND:', cc_OND)
print('Correlation Coefficient of SON:', cc_SON)
print('Correlation Coefficient of ASO:', cc_ASO)
print('Correlation Coefficient of JAS:', cc_JAS)
print('Correlation Coefficient of JJA:', cc_JJA)

max_cc =  max(cc_NDJ, cc_OND, cc_SON, cc_ASO, cc_JAS, cc_JJA)
print('Maximum Correlation Coefficient : ', max_cc)
```
