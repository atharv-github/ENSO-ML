```python
import matplotlib.pyplot as plt
import pandas as pd

x = cleaned_data['Time']
y = cleaned_data['ENSO val']

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(x, y, color='black', linewidth=1)

ax.fill_between(x, y, where=(y > -0.4)  , color='white', alpha=0.3, label='Neutral')
ax.fill_between(x, y, where=(y > 0.4), color='red', alpha=0.3, label='El Nino')
ax.fill_between(x, y, where=(y < -0.4), color='blue', alpha=0.3, label='La Nina')

ax.axhline(0, color='black', linewidth=0.5)

ax.set_xlabel('Time')
ax.set_ylabel('ENSO val')
ax.set_title('ENSO Values Over Time')

ax.legend(title='Conditions')

fig.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

df=pd.read_csv('datacleaned.csv')

el_nino_df = df[df['ENSO val'] > 0.4]

la_nina_df = df[df['ENSO val'] < -0.4]

plt.figure(figsize=(10, 6))
plt.plot(el_nino_df['Time'], el_nino_df['ENSO val'], color='red', label='El Niño')
plt.xlabel('Time')
plt.ylabel('ENSO Value')
plt.title('El Niño Trends')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(la_nina_df['Time'], la_nina_df['ENSO val'], color='blue', label='La Niña')
plt.xlabel('Time')
plt.ylabel('ENSO Value')
plt.title('La Niña Trends')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

el_nino_df = df[df['ENSO val'] > 0.5]

la_nina_df = df[df['ENSO val'] < -0.5]

plt.scatter(el_nino_df.index, el_nino_df['ENSO val'], c='red', label='El Niño')

plt.scatter(la_nina_df.index, la_nina_df['ENSO val'], c='blue', label='La Niña')

plt.title('ENSO Values - El Niño and La Niña')
plt.xlabel('Time')
plt.ylabel('ENSO Value')
plt.legend()
plt.figure(figsize=(10, 6))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_values = np.arange(len(df))

el_nino_df = df[df['ENSO val'] > 0.4]

la_nina_df = df[df['ENSO val'] < -0.4]

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x_values, df['ENSO val'], label='Neutral')
plt.scatter(x_values[el_nino_df.index], el_nino_df['ENSO val'], color='r', label='El Niño')
plt.scatter(x_values[la_nina_df.index], la_nina_df['ENSO val'], color='b', label='La Niña')
plt.title('ENSO Values over Time')
plt.xlabel('Time')
plt.ylabel('ENSO Value')
plt.xticks(x_values[::12], df['Time'][::12], rotation=45)
plt.legend()
plt.show()

# Line plot
plt.figure(figsize=(10, 6))
plt.plot(x_values, df['ENSO val'], label='Neutral')
plt.plot(x_values[el_nino_df.index], el_nino_df['ENSO val'], 'r', label='El Niño')
plt.plot(x_values[la_nina_df.index], la_nina_df['ENSO val'], 'b', label='La Niña')
plt.title('ENSO Values over Time')
plt.xlabel('Time')
plt.ylabel('ENSO Value')
plt.xticks(x_values[::12], df['Time'][::12], rotation=45)
plt.legend()
plt.show()

# Area plot
plt.figure(figsize=(10, 6))
plt.fill_between(x_values, df['ENSO val'], alpha=0.5, label='Neutral')
plt.fill_between(x_values[el_nino_df.index], el_nino_df['ENSO val'], alpha=0.5, color='r', label='El Niño')
plt.fill_between(x_values[la_nina_df.index], la_nina_df['ENSO val'], alpha=0.5, color='b', label='La Niña')
plt.title('ENSO Values over Time')
plt.xlabel('Time')
plt.ylabel('ENSO Value')
plt.xticks(x_values[::12], df['Time'][::12], rotation=45)
plt.legend()
plt.show()
```
